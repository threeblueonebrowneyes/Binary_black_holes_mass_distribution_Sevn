```python
import re

# Dataframe for NCE
regex_str = "B;(?:\d*\_)?\d+;(\d+);CE;"

logfile_path = 'sevn_output_Z0.01A1L1/sevn_output_Z0.01A1L1/0/logfile_0.dat'
with open(logfile_path,"r") as fo:
    i=fo.read()
    ma=re.findall(regex_str, i)
    
NCE = pd.DataFrame({'ID':np.asarray(ma,dtype=int)})
NCE = NCE.groupby(['ID']).size().reset_index(name='NCE')

#Dataframe for NRLO
regex_str = "B;(?:\d*\_)?\d+;(\d+);RLO_BEGIN;"

logfile_path = 'sevn_output_Z0.01A1L1/sevn_output_Z0.01A1L1/0/logfile_0.dat'
with open(logfile_path,"r") as fo:
    i=fo.read()
    ma=re.findall(regex_str, i)
    
NRLO = pd.DataFrame({'ID':np.asarray(ma,dtype=int)})
NRLO = NRLO.groupby(['ID']).size().reset_index(name='NRLO')
ma
```
